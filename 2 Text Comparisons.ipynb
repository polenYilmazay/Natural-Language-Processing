```python
import nltk ,collections
from nltk.util import ngrams
from nltk.collocations import *
from collections import Counter
```

```python
file = open('Liza_and_the_Lost_Letter_by_Bruce_Lansky.txt',encoding="utf-8")
file=file.read()

stopwords = nltk.corpus.stopwords.words('English') 
tokenized = file.split()
tokenized = [token.lower() for token in tokenized if token.isalpha()]
```

```python
Unigrams=ngrams(tokenized,1)
UnigramsFreq=collections.Counter(Unigrams)
UnigramsFreq.most_common(10)
#Eşik değeri istenirse
valuesOfUnigrams=list(UnigramsFreq.values())
unigramlist=list(UnigramsFreq)

for x in range(0,len(unigramlist)):
    if(valuesOfUnigrams[x]>30):
        print(unigramlist[x],"kelimesi",valuesOfUnigrams[x],"kez kullanıldı")
```

```python
#Bigrams=ngrams(tokenized,2)
#BigramsFreq=collections.Counter(Bigrams)
#BigramsFreq.most_common(10)

#Belirli bir eşik değeri istersek eğer
bigram= nltk.collocations.BigramAssocMeasures()
finder2 = BigramCollocationFinder.from_words(tokenized)
finder2.apply_freq_filter(10)
print("Bigram :")

sorted(finder2.ngram_fd.items(),key=lambda t:(-t[1],t[0]))[:10]
```

```python
#Trigrams=ngrams(tokenized,3)
#TrigramsFreq=collections.Counter(Trigrams)
#TrigramsFreq.most_common(10)
trigram=nltk.collocations.TrigramAssocMeasures
finder3=TrigramCollocationFinder.from_words(tokenized)
finder3.apply_freq_filter(4) #eşik değeri
print("Trigram :")
sorted(finder3.ngram_fd.items(),key=lambda t:(-t[1],t[0]))[:10]
```

```python
tag1=nltk.pos_tag(tokenized)
print(tag1)


counts=Counter(tag for tokenized, tag in tag1)
print("\n",counts)
```
