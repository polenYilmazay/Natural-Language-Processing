```python
import nltk
from nltk.stem import PorterStemmer, SnowballStemmer
from snowballstemmer import TurkishStemmer 
```

```python
#Porter Stemmer ile kök bulma. Stemmer sadece ekleri çıkartır ve geriye ne 
#ne kalırsa onu bize sunar. Bağlama göre hareket etmez.
porter=PorterStemmer()

word_list=["Study","Studies","Studying","Studied"]

for w in word_list:
    print(porter.stem(w))
       
```

```python
word_list=["studies","leaves","decreases","plays"]

for w in word_list:
    print(porter.stem(w))
```

```python
snowball=SnowballStemmer("english")

word_list=["Study","Studies","Studying","Studied"]


for w in word_list:
    print(snowball.stem(w))
```

```python
from snowballstemmer import stemmer
turkStem=stemmer('turkish')
turkStem.stemWord("ekmekler")
```

```python
file=open("Mustafa_Kemal_Atatürk.txt",encoding="utf-8")
file=file.read()
tokenized = file.split()
tokenized = [token.lower() for token in tokenized if token.isalpha()]
```

```python
data_words_nostops = [turkStem.stemWord(word) for word in tokenized]
print(data_words_nostops)
```
