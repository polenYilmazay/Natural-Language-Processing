```python
TurkishMorphology: JClass = JClass('zemberek.morphology.TurkishMorphology')

morphology: TurkishMorphology = TurkishMorphology.createWithDefaults()

#text: str = 'Truva, Küçük Asya denen Anadolu/nun kuzeybatısındaki Troas bölgesinde bir sırtın üzerinde, Çanakkale/nin 30 km. kadar uzağındaki Hisarlık Tepesi üzerinde, dokuz kere yıkılıp yeniden kurulmuş olan bir şehirdir'

data=open("Truva_Savası.txt",encoding="utf-8")
text: str=data.read()

analysis: java.util.ArrayList = (
        morphology.analyzeAndDisambiguate(text).bestAnalysis()
    )

pos: List[str] = []

for i, analysis in enumerate(analysis, start=1):
        print(
            f'\nAnalysis {i}: {analysis}',
            f'\nPrimary POS {i}: {analysis.getPos()}'
            #f'\nPrimary POS (Short Form) {i}: {analysis.getPos().shortForm}'
        )
        pos.append(
            f'{str(analysis.getLemmas()[0])}'
            f'-{analysis.getPos().shortForm}'
        )

#print(f'\nFull sentence with POS tags: {" ".join(pos)}')
```
